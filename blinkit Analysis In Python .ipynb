## Cleaning data

```python
df['Item Fat Content']= df['Item Fat Content'].replace({'reg':'Regular','low fat':'Low Fat','lf':'Low Fat','LF':'Low Fat'})
```

```python
df['Item Fat Content'].unique()
```

# Business Requirement Analysis

## KPI

```python
Total_sale=df['Sales'].sum()
print(f"Total Sales is :${Total_sale:,.0f}")

Average_Sale=df['Sales'].mean()
print(f"Average sales is:${Average_Sale:,.0f}")

Average_Rating=df['Rating'].mean()
print(f"Average Rating:{Average_Rating:.0f}")

Number_Of_Item=df['Sales'].count()
print(f"Number of item is:{Number_Of_Item}")
```

# Charts Requirements

## Total Sales By Fat Content

```python
sale_fat=df.groupby('Item Fat Content')['Sales'].sum()
plt.pie(sale_fat,labels=sale_fat.index,autopct='%.1f%%',startangle=90)
plt.title('Sales By Fat Content')
```

## Total Sales By Item

```python
Item_Sales=df.groupby('Item Type')['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='Item Type',y='Sales',data=Item_Sales)
plt.title('Total Sales By Item')
plt.tight_layout()
plt.xticks(rotation=-90)
plt.show()
```

## Total Sales by outlet Establishment

```python
outlet_establishment=df.groupby('Outlet Establishment Year')['Sales'].sum()
```

```python
plt.figure(figsize=(10,6))
plt.plot(outlet_establishment.index,outlet_establishment.values,marker='o',linestyle='-')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Sales')
plt.title('Outlet Establishment')
plt.tight_layout()
plt.show()
```

## Fat content by outlet for Total Sales

```python
grouped=df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
grouped=grouped[['Regular','Low Fat']]
ax=grouped.plot(kind='bar',figsize=(10,8),title='Fat content by outlet for Total Sales')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

## Sales By Outlet Size

```python
Sales_Outlet_Size=df.groupby('Outlet Size')['Sales'].sum()
plt.figure(figsize=(10,6))
plt.pie(Sales_Outlet_Size,labels=Sales_Outlet_Size.index,autopct='%.1f%%',startangle=90)
plt.title('Sales By Outlet Size')
plt.tight_layout()
plt.show()
```

## Total Sales By Outlet Location Type

```python
Sales_Location_Type=df.groupby('Outlet Location Type')['Sales'].sum().reset_index()

plt.figure(figsize=(10,6))
xs=sns.barplot(x='Sales',y='Outlet Location Type',data=Sales_Location_Type)
plt.tight_layout()
plt.show()
```
